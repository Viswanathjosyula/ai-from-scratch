# Classic algorithms — does the simple model really win? (NYC taxi fares)

In Module 1 we used a decision tree to predict taxi fares. Here we put **four different
algorithms** head-to-head on the *same* real NYC taxi data and ask one question every good
ML engineer asks first:

**Does a plain, simple model beat the fancy ones — or not?**

The four contenders, simplest to fanciest:
- **Linear Regression** — draws the best straight-line relationship. Decades old. The baseline.
- **Decision Tree** — a flowchart of yes/no splits.
- **Random Forest** — hundreds of trees voting (this is *bagging*).
- **Gradient Boosting** — trees built in sequence, each fixing the last's mistakes (this is *boosting*).

Press play on each cell. ▶


```python
# Real NYC taxi trips — built into seaborn, no download
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor

df = sns.load_dataset("taxis").dropna(subset=["distance", "passengers", "fare"])

X = df[["distance", "passengers"]]   # the trip
y = df["fare"]                       # predict the fare (a number)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(f"{len(X_train)} trips to train on, {len(X_test)} held out to test")
```

### Train all four on the same data, score them on rides they've never seen

We score with **R²** (1.0 = perfect, 0.0 = no better than guessing the average fare).
The number that matters is the **test** R² — performance on new, unseen rides.


```python
models = {
    "Linear Regression": LinearRegression(),
    "Decision Tree":     DecisionTreeRegressor(random_state=42),
    "Random Forest":     RandomForestRegressor(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
}

print(f"{'model':>20} | {'train R2':>8} | {'test R2':>8}")
print("-" * 44)
for name, m in models.items():
    m.fit(X_train, y_train)
    print(f"{name:>20} | {m.score(X_train, y_train):>8.3f} | {m.score(X_test, y_test):>8.3f}")
```

### What to notice

Two things almost always show up:

1. **The single Decision Tree overfits** — high train R², lower test R². (Same trap as Module 1.)
2. **The simple Linear Regression is competitive** — often within a hair of the fancy ensembles,
   sometimes beating them.

That's the real lesson: a senior engineer reaches for the simple baseline *first*. It's cheaper,
faster, easier to explain — and surprisingly hard to beat. Fancy is never free; you only pay for
it when it actually earns its keep.
